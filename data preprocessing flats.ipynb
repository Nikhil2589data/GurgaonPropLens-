```python
# to remove rating from the society name using re(regular expression module)
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
# price column
df['price'].value_counts()
```

```python
# some flats have price on request
df[df['price'] == 'Price on Request']
```

```python
# removing the columns which have price = price on request 
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
# function to convert prices into a single quantity i.e. is crores
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python

df.sample(5)
```

```python

# price_per_sqft
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'] = (
    df['price_per_sqft']
    .astype(str)                              # Convert to string
    .str.split('/')
    .str.get(0)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
    .replace('nan', None)                     # Convert string 'nan' to real NaN
    .astype(float)                            # Finally convert to float
)
```
